# Classification Based Detection

```python
# Config
image_dir = '/content/drive/MyDrive/Upload to Google Colab/satellite_imagery_cropped'

# CSV with columns: filename, has_solar (1 or 0)
labels_csv = '/content/drive/MyDrive/Upload to Google Colab/solar_panel_presence_labels.csv'
conf_threshold = 0.25
```

```python
# Load model
model = YOLO((FILE_PATH+'/Small Model/best.onnx'), task='detect')

# Load labels
df = pd.read_csv(labels_csv)
results = []
df
```

```python
# Loop through images
for _, row in df.iterrows():
    print('image:',_)
    file = row['filename']
    label = row['has_solar']
    img_path = Path(image_dir) / file

    # Run prediction
    pred = model(img_path, conf=conf_threshold)[0]
    detected = len(pred.boxes) > 0
    results.append({
        'filename': file,
        'actual': label,
        'predicted': int(detected)
    })
```

```python
# Create dataframe
results_df = pd.DataFrame(results)
```

```python
results_df
```

```python
# Evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix

y_true = results_df['actual']
y_pred = results_df['predicted']
```

```python
print(confusion_matrix(y_true, y_pred))
print(classification_report(y_true, y_pred, target_names=["No Solar", "Has Solar"]))
```
